```r
# Importing the haven and ggplot libraries 
library(haven)
library(ggplot2)
library(car)
```

```r
# Importing the necessary data file
df_protest <- read_spss('Data protest.sav')
```

```r
df_protest
```

```r
str(df_protest)
```

```r
df_protest_scaled <- df_protest
df_protest_scaled$sexism <- scale(df_protest_scaled$sexism)
df_protest_scaled$liking <- scale(df_protest_scaled$liking)
df_protest_scaled$respappr <- scale(df_protest_scaled$respappr)
```

```r
df_protest_scaled$protest <- as.factor(df_protest_scaled$protest)
```

```r
ggplot(data = df_protest_scaled, aes(x = sexism, y = liking, color = protest)) + geom_point() + geom_smooth(method = 'lm', se = FALSE) + labs(title = 'Sexism versus Liking of Subject')
```

```r
ggplot(data = df_protest_scaled, aes(x = respappr, y = liking, color = protest)) + geom_point() + geom_smooth(method = 'lm', se = FALSE) + labs(title = 'Sexism versus Approval of Response')
```

```r
df_protest_scaled
```

```r
df_protest_scaled$protest <- as.numeric(df_protest_scaled$protest) - 1
df_protest_scaled$protest_sexism <- df_protest_scaled$protest * df_protest_scaled$sexism
df_protest_scaled$protest <- as.factor(df_protest_scaled$protest)
```

```r
df_protest_scaled
```

```r
liking_model <- lm(liking ~ sexism + protest + protest_sexism, data = df_protest_scaled)
```

```r
summary(liking_model)
```

```r
approval_model <- lm(respappr ~ sexism + protest + protest_sexism, data = df_protest_scaled)
```

```r
summary(approval_model)
```
